```python
import pandas as pd
from collections import Counter
import re

# Leer el archivo CSV
df = pd.read_csv('C:/Users/mafer/Downloads/data.csv')

# Recorre todas las filas del DataFrame
for fila in range(len(df)):
    # Reiniciar las variables al principio de cada iteración
    blue = False
    color2 = None
    color3 = None
    color4 = None
    sparkle = False
    shape = ""
    output = ""
    
    compania = df.at[fila, 'Compania']
    
    # Determinar el tipo de gráfico según la compañía
    if compania == "amigos" or compania == "familia":
        shape = "Group graph"
    elif compania == "pareja":
        shape = "Couple graph"
    else:
        shape = "Simple graph"
    
    viaje = df.at[fila, 'Viaje']
    if viaje == "Si":
        blue = True
    
    naturaleza = df.at[fila, 'Naturaleza']
    if naturaleza == 1:
        color2 = "green"
    
    logro = df.at[fila, 'logro']
    if logro == 1:
        color3 = "pink"
    
    fiesta = df.at[fila, 'fiesta']
    if fiesta == 1:
        color4 = "orange"
    
    sonrisa = df.at[fila, 'sonrisa']
    if sonrisa == 1:
        sparkle = True
    
    space = df.at[fila, 'Interior/Exterior']   
    ubi = df.at[fila, 'Ubicacion / Location']
    

    
    # Construir la salida
    output = shape

    if blue == True:
        output += " " + "blue"
    if color2:
        output += " " + color2
    if color3:
        output += " " + color3
    if color4:
        output += " " + color4
    if space == "ext":
        output += " add contour"
    if sparkle == True:
        output += " with sparkle!"
    
    # Imprimir el resultado de esta fila
    print(output)
```
